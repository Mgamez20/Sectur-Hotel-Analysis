```python

def occupied_sum_quintanaroo(lista2019, lista2020, loc):
    #locations = ['Cozumel', 'Cancún', 'Isla Mujeres', 'Playa del Carmen']
    stars = ['1 star', '2 stars', '3 stars', '4 stars', '5 stars']
    for star in stars:
        occupied_sumstar_2019 = occupied_rooms_stars_quintana_roo_2019[(occupied_rooms_stars_quintana_roo_2019['location'] == loc)&(occupied_rooms_stars_quintana_roo_2019['hotel_rating'] == star)].sum()[1]
        occupied_sumstar_2020 = occupied_rooms_stars_quintana_roo_2020[(occupied_rooms_stars_quintana_roo_2020['location'] == loc)&(occupied_rooms_stars_quintana_roo_2020['hotel_rating'] == star)].sum()[1]
        lista2019.append(occupied_sumstar_2019)
        lista2020.append(occupied_sumstar_2020)
    
    return lista2019, lista2020
```

```python
#Sum occupied rooms star 1-5 Cozumel 2019-2020
list_cancun_2019, list_cancun_2020 = [], []

list_cancun_2019, list_cancun_2020 = occupied_sum_quintanaroo(list_cancun_2019, list_cancun_2020, 'Cozumel')

print(list_cancun_2019)
print(list_cancun_2020)

df_quintanaroo_top_locations = pd.DataFrame({"Stars": [1,2,3,4,5],
                                            "Cozumel 2019":list_cancun_2019,
                                            "Cozumel 2020": list_cancun_2020})

df_quintanaroo_top_locations
```

```python
#filtering total nights by tourists 2019-2020
sum_2019_tourist_quintanaroo = df_2019_tourist[df_2019_tourist['geography'] == 'Quintana Roo'].sum()[1]
sum_2020_tourist_quintanaroo = df_2020_tourist[df_2020_tourist['geography'] == 'Quintana Roo'].sum()[1]

sum_2019_tourist_chiapas = df_2019_tourist[df_2019_tourist['geography'] == 'Chiapas'].sum()[1]
sum_2020_tourist_chiapas = df_2020_tourist[df_2020_tourist['geography'] == 'Chiapas'].sum()[1]

sum_2019_tourist_bcsur = df_2019_tourist[df_2019_tourist['geography'] == 'Baja California Sur'].sum()[1]
sum_2020_tourist_bcsur = df_2020_tourist[df_2020_tourist['geography'] == 'Baja California Sur'].sum()[1]

sum_2019_tourist_bc = df_2019_tourist[df_2019_tourist['geography'] == 'Baja California'].sum()[1]
sum_2020_tourist_bc = df_2020_tourist[df_2020_tourist['geography'] == 'Baja California'].sum()[1]
```

```python
#filtering total nights by tourists 2019-2020
sum_2019_arrivals_quintanaroo = df_2019_arrivals[df_2019_arrivals['geography'] == 'Quintana Roo'].sum()[1]
sum_2020_arrivals_quintanaroo = df_2020_arrivals[df_2020_arrivals['geography'] == 'Quintana Roo'].sum()[1]

sum_2019_arrivals_chiapas = df_2019_arrivals[df_2019_arrivals['geography'] == 'Chiapas'].sum()[1]
sum_2020_arrivals_chiapas = df_2020_arrivals[df_2020_arrivals['geography'] == 'Chiapas'].sum()[1]

sum_2019_arrivals_bcsur = df_2019_arrivals[df_2019_arrivals['geography'] == 'Baja California Sur'].sum()[1]
sum_2020_arrivals_bcsur = df_2020_arrivals[df_2020_arrivals['geography'] == 'Baja California Sur'].sum()[1]

sum_2019_arrivals_bc = df_2019_arrivals[df_2019_arrivals['geography'] == 'Baja California'].sum()[1]
sum_2020_arrivals_bc = df_2020_arrivals[df_2020_arrivals['geography'] == 'Baja California'].sum()[1]
```

```python
def avg_stay(tourist, arrivals):
    avgstay = tourist / arrivals
    return avgstay
```

```python
avg_stay_2019_quintanaroo = avg_stay(sum_2019_tourist_quintanaroo, sum_2019_arrivals_quintanaroo)
avg_stay_2020_quintanaroo = avg_stay(sum_2020_tourist_quintanaroo, sum_2020_arrivals_quintanaroo)

avg_stay_2019_chiapas = avg_stay(sum_2019_tourist_chiapas, sum_2019_arrivals_chiapas)
avg_stay_2020_chiapas = avg_stay(sum_2020_tourist_chiapas, sum_2020_arrivals_chiapas)

avg_stay_2019_bcsur = avg_stay(sum_2019_tourist_bcsur, sum_2019_arrivals_bcsur)
avg_stay_2020_bcsur = avg_stay(sum_2020_tourist_bcsur, sum_2020_arrivals_bcsur)

avg_stay_2019_bc = avg_stay(sum_2019_tourist_bc, sum_2019_arrivals_bc)
avg_stay_2020_bc = avg_stay(sum_2020_tourist_bc, sum_2020_arrivals_bc)
```

```python
avg_stay_2019 = [avg_stay_2019_quintanaroo, avg_stay_2019_chiapas, avg_stay_2019_bcsur, avg_stay_2019_bc]
avg_stay_2020 = [avg_stay_2020_quintanaroo, avg_stay_2020_chiapas, avg_stay_2020_bcsur, avg_stay_2020_bc]

# Create a sample dataframe
df = pd.DataFrame({'2019': avg_stay_2019,
                   '2020': avg_stay_2020},
                  index=['Quintana Roo', 'Chiapas', 'Baja California Sur', 'Baja California'])
  
# Plot grouped horizontal bar chart

df.plot.barh(title="Average lenght of stay in the most visited states by foreing tourists between 2019 and 2020",
             color={'#FFDB15', '#3F5E98'}, 
            figsize = (10,10))
plt.xlabel('Days', Fontsize = 12)
plt.ylabel('Most visited states', Fontsize = 12)
```
